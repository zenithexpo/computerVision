```python
import matplotlib.image as matimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

```python
#reading the image
image = matimg.imread('diksha.jpg')
```

```python
#printing information about image
print("dimensions are:", image.shape)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray_image, cmap='gray')
```

```python
#getting value of the pixel at specific cooridinates
print(gray_image[1000,1750])
```

```python
#printing maximum and minimum value of the pixel in an image
print(np.amax(gray_image),np.amin(gray_image))
```

```python
#creating 3X3 image with grey scale
tiny_image=np.array([[3,255,67],[34,78,93],[100,73,60]])
plt.matshow(tiny_image)
```

```python
#RGB channels of the image
r=image[:,:,0]
g=image[:,:,1]
b=image[:,:,2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))
ax1.set_title('R channel')
ax1.imshow(r, cmap='gray')
ax2.set_title('G channel')
ax2.imshow(g, cmap='gray')
ax3.set_title('B channel')
ax3.imshow(b, cmap='gray')
```
