## Images as Numerical Data
Images are just 2D representation of 3D world scenes. They grid of pixels each having a value in range of 0-225. 

Important resources (libraries) are imported below.

```
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

Read in and Display the Image

```
# Read in the image
image = cv2.imread("/content/drive/My Drive/photos/IMG-20180904-WA0012.jpg")
# Display the image dimensions
print("Image dimensions:", image.shape)

# Change from color to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Output the gray image
plt.imshow(gray_image, cmap='gray') 


```

```
#Let us try some hands on this image
#print specific grayscale pixel values
print(gray_image[400,400])
```

Syntax: gray_image[y,x] <br>
Computer looks images as a numpy array! with values in them, so you can simply print the value and treat them as a array.

```
# Print maximum and minimum value in gray image
max_val = np.amax(gray_image)
min_val = np.amin(gray_image)
print(max_val)
print(min_val)
```

## Reverse It!
For a machine, images are just numerical data i.e Grids of number as an input!<br>
Now, Give input as grid of numbers and see what's the image we have got.

```
# Create a 5x5 image using just grayscale, numerical values
tiny_image = np.array([[250, 250, 250, 250, 250],
                      [250, 0, 250, 0, 250],
                      [250, 250, 250, 250, 250],
                      [250, 50, 250, 50, 250],
                      [250, 250, 50, 250, 250]])

# To show the pixel grid, use matshow
plt.matshow(tiny_image, cmap='gray')
```
